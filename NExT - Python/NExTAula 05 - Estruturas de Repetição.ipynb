# Estruturas de Repeti√ß√£o

Em Python existem apenas duas estruturas de repeti√ß√£o:
- while;
- for.

Essas estruturas s√£o capazes de repetir um bloco de c√≥digo at√© que determinada consid√ß√£o n√£o seja mais atendida ou uma estrutura de dados seja percorrida.

üí° *A pesar de s√≥ existirem essas estruturas de repeti√ß√£o, o Python fornece outros comandos que s√£o capazes de iterar em listas e fazer manipula√ß√µes de seus elementos de forma pr√°tica.*

√â muito comum usarmos estruturas de repeti√ß√£o junto a lista.

Ex.: Exibir todos os elementos de uma lista sem uma estrutura de repeti√ß√£o:

````python
lista = ['item1', 'item2', 'item3', 'item4', 'item5']

print(lista[0] + ' √© um item dessa lista')
print(lista[1] + ' √© um item dessa lista')
print(lista[2] + ' √© um item dessa lista')
print(lista[3] + ' √© um item dessa lista')
print(lista[4] + ' √© um item dessa lista')
````

# üîÅ While

Essa estrutura permite repetir um trecho de c√≥digo enquanto uma condi√ß√£o for verdadeira (`True`)

```python
while condi√ß√£o:
  # c√≥digo que vai repetir
```

‚ö†Ô∏è Aten√ß√£o para a identa√ß√£o dos blocos!

````python
# repetir at√© que a pessoa acerte a senha
senha = '123456'

senha_usuario = input('Insira sua senha: ')

while senha_usuario != senha:
  print('Voc√™ errou sua senha, seu boc√≥\nTente novamente')
  senha_usuario = input('Insira sua senha')

print('Finalmente, achei que voc√™ nunca ia acertar üëç')
````

````python
# percorrer todos os elementos de uma lista
estudantes = ['Bruno', 'Dustin', 'Anna', 'Shell', 'Diego']

i = 0
while i < len(estudantes):
  print(f'Seja bem-vindo(a) {estudantes[i]}')
  i += 1
````

````python
# Exibir um n√∫mero x, at√© que a soma de todos os n√∫meros exibidos seja 100
x = somatorio = 0

while somatorio <= 100:
  x += 1
  somatorio += x
  print(x, somatorio)
````

````python
# else

x, somatorio = 0, 0

while somatorio <= 100:
  x += 1
  somatorio += x
  print(x, somatorio)
  #break
else:
  print('Loop encerrado com sucesso üíòü¶≠')
````

# üîÉ `For`

Essa estrutura √© um iterador, ou seja, ele percorre itens que est√£o em sequ√™ncia, tais como listas ou strings.



```python
for item in objeto:
  # c√≥digo que vai repetir
```



````python
# percorrer uma lista
estudantes = ['Bruno', 'Dustin', 'Anna', 'Shell', 'Diego']

'''para(inteiro i=0; i<5; i++) {
    escreva(estudantes[i])
}'''

for aluno in estudantes:
  print(f'Nome do estudante: {aluno}')
````

````python
# range

for num in range(10, 50):
  print(num)

#print(list(range(10, 50)))
````

````python
estudantes = ['Bruno', 'Dustin', 'Anna', 'Shell', 'Diego']

for i in range(len(estudantes)):
  print(i, estudantes[i])
````

````python
# exibir todos os n√∫meros pares contidos de 1 a 100

for num in range(1, 101):
  if num % 2 == 0:
    print(num)
````

````python
# exibir todos os n√∫meros pares contidos de 1 a 100

for num in range(2, 101, 2):
  print(num)
````

````python
# string

nome = input()

for letra in nome:
  print(ord(letra))
````

# üîÄ Controle de Fluxo

- `break`: Para um loop
- `continue`: Vai para o pr√≥ximo loop
- `pass`: n√£o faz nada



```python
while condi√ß√£o:
  # c√≥digo
  if teste1:
    break
  if teste2:
    continue
  else:
    pass
else:
  pass
```



````python
i = 10

if i < 100:
  # continuar depois...
  pass

print('Olha a foca ü¶≠')
````

````python
# interromper um loop quando ler um n√∫mero negativo

while True:
  num = int(input('Insira um n√∫mero: '))

  if num < 0:
    break
else:
  print('Loop finalizado')
````

````python
# somar 10 n√∫meros, a n√£o ser que seja negativo

i = somatorio = 0
while i < 10:
  num = int(input('Insira um n√∫mero: '))

  if num >= 0:
    somatorio += num
  else:
    break

  i += 1
else:
  print('Todos os 10 n√∫meros foram lidos ‚úåÔ∏è')

print(f'Valor do somat√≥rio = {somatorio}')
````

````python
i = somatorio = 0
while i < 10:
  num = int(input('Insira um n√∫mero: '))

  if num >= 0:
    somatorio += num
  else:
    continue

  if somatorio > 20:
    break

  i += 1
else:
  print('Todos os 10 n√∫meros foram lidos ‚úåÔ∏è')

print(f'Valor do somat√≥rio = {somatorio}')
````

7. Encontrar n√∫meros primos √© uma tarefa dif√≠cil. Fa√ßa um programa que gera uma lista dos n√∫meros primos existentes entre 1 e um n√∫mero inteiro informado pelo usu√°rio.

````python

numero = int(input())

primos = []

for i in range(2,numero):
  n_divisores = 0
  for j in range(2,i):
    if i % j == 0:
      n_divisores = 1
      break;

  if n_divisores == 0:
    primos.append(i)
  

print(primos)
````
