# üì¶ Vari√°veis

Em termos t√©cnicos, uma vari√°vel atribui um local de armazenamento a um valor que est√° vinculado a um nome ou identificador simb√≥lico. O nome da vari√°vel √© usado para fazer refer√™ncia a esse valor armazenado em um programa de computador.

Python √© uma linguagem de tipagem **din√¢mica** e **forte**:
*   **Tipagem Din√¢mica**: o interpretador Python infere qual o tipo do dado que a vari√°vel est√° recebendo (n√£o √© necess√°rio inserir explicitamente);
*   **Tipagem Forte**: o interpretador Python **n√£o** faz coer√ß√µes autom√°ticas (convers√µes de valores) entre tipos n√£o compat√≠veis.

Desta forma, para criar uma vari√°vel em Python, basta declarar seu nome e atribuir um valor.

‚ö†Ô∏è Em Python, n√£o √© poss√≠vel declarar uma vari√°vel sem atribuir um valor a ela

# üßÆ Opera√ß√µes Aritm√©ticas

Em Python as 4 opera√ß√µes aritm√©ticas b√°sicas s√£o feitas usando os operadores tradicionais `+`, `-`, `*` e `/`.

- `x + y`:	soma de x e y
- `x - y`:	subtra√ß√£o de x e y
- `x * y`:	multiplica√ß√£o de x e y
- `x / y`:	divis√£o de x e y

üí° Qualquer divis√£o usando o operador `/`, mesmo que usando n√∫meros `int`, gera um resultado em `float`

Os operadores de `+`  e `-` tamb√©m podem ser usados para manipular o sinal do n√∫mero:

- `+x`:	identidade de x
- `-x`:	sinal alterado de x

Python tamb√©m disp√µe de outros operadores :

- `x % y`:	resto da divis√£o de x por y
- `x // y`:	divis√£o **inteira** de x e y
- `x ** y`:	x elevado a pot√™ncia de y (`x ∏`)


````
print(10/2)
print(10//2)
print(3**.5)
````

## Preced√™ncia Aritm√©tica

A ordem das opera√ß√µes aritm√©ticas em Python √© semelhante a matem√°tica e pode ser modificada com o uso de par√™nteses `()`

A preced√™ncia aritm√©tica em Python segue a ordem:
1. `()`
2. `**`
3. `*`, `/` (*left-associative*)
4. `+`, `-` (*left-associative*)

```python
2 + 10 * 10 - 5 # 97
(2 + 10) * (10 - 5) # 60
```

````
print(2 + 10 * 10 - 5)
print((2 + 10) * (10 - 5))
````

## Operadores de atribui√ß√£o


* `+=` - - adiciona e atribui o valor
* `-=` - - subtrai e atribui o valor
* `*=` - - multiplica e atribui o valor
* `/=` - - divide e atribui o valor
* `//=` - - divide e depois atribuir o valor em inteiro
* `**=` - - aumentar a pot√™ncia e ent√£o atribuir o valor
* `%=` - - retorna o resto da divis√£o e depois atribui o valor

üí° N√£o existe `++` ou `--` em Python „Ñü( ‚ñî, ‚ñî )„Ñè

15. Fa√ßa um Programa que pergunte quanto voc√™ ganha por hora e o n√∫mero de horas trabalhadas no m√™s. Calcule e mostre o total do seu sal√°rio no referido m√™s, sabendo-se que s√£o descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, fa√ßa um programa que nos d√™:
    - sal√°rio bruto.
    - quanto pagou ao INSS.
    - quanto pagou ao sindicato.
    - o sal√°rio l√≠quido.
    - calcule os descontos e o sal√°rio l√≠quido, conforme a tabela abaixo:

```
            + Sal√°rio Bruto : R$
            - IR (11%) : R$
            - INSS (8%) : R$
            - Sindicato ( 5%) : R$
            = Sal√°rio Liquido : R$
```
> Obs.: Sal√°rio Bruto - Descontos = Sal√°rio L√≠quido.


````
valorHora = float(input('Quanto voce ganha por hora? R$ '))
numHoras = float(input('Quantas horas voce trabalha por m√™s? '))

salarioBruto = valorHora * numHoras
IR = salarioBruto * 0.11 
Inss = salarioBruto * 0.08
Sindicato = salarioBruto * 0.05
descontos = IR + Inss + Sindicato
salarioLiquido = salarioBruto - descontos

print('Sal√°rio bruto => R$ ', salarioBruto)
print('IR => R$ ', IR )
print('INSS => R$ ', Inss)
print('Sindicato => R$ ', Sindicato)
print('Salario Liquido => R$ ', salarioLiquido)

````
