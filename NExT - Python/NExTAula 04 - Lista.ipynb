# üßæ Listas

Quando desenvolvemos um sistema, √© comum que seja necess√°rio trabalhar com uma grande quantidade de dados semelhantes que podem variar de quantidade. Ex.:
- Lista de amigos em uma rede social;
- Lista de itens adicionados a um carrinho de compras em um site;
- Lista de convidados em uma reuni√£o.

Para esses casos, criar uma vari√°vel para cada item √© invi√°vel. A solu√ß√£o ent√£o √© trabalhar com uma estrutura de dados que seja capaz de armazenar uma quantidade vari√°vel de informa√ß√µes. A essas estruturas chamamos de **lista** (`list`).

Nesses casos, a quantidade n√£o √© o mais importante, mas sim a facilidade de ter agrupado valores de um mesmo tipo.

> Cole√ß√µes de itens em uma **ordem** espec√≠fica.

# Manipula√ß√£o de listas

Diferente das *strings* que s√£o cole√ß√µes imut√°veis, listas s√£o **din√¢micas** e podem aumentar e diminutir de tamanho e ainda ter seus elementos alterados.

## Alterando um elemento de uma lista

Para alterar um elemento de uma lista, basta atraibuir um novo valor a uma posi√ß√£o j√° existente

````python
lista_de_chamada = ['fabricio', 'erick', 'augusto','julio']
print(lista_de_chamada)
lista_de_chamada[-1] = 'nicole'
print(lista_de_chamada)
````

## ‚ûï Adicionando elementos em uma lista

√â poss√≠vel adicionar novos elementos em um objeto `list` j√° existente usando:

```python
l.append(elem)
l.insert(elem)
l.extend(lista)
```


### Adicionando ao final da lista

`l.append(elem)`

````python
lista_de_chamada = ['fabricio', 'erick', 'augusto','julio']
print(lista_de_chamada)
lista_de_chamada.append('nicole')
print(lista_de_chamada)
````

### Adicionando em qualquer posi√ß√£o

`insert(pos, elem)`

````python
lista_de_chamada = ['fabricio', 'erick', 'augusto','julio']
#                       0         1         2          3
print(lista_de_chamada)

lista_de_chamada.insert(1, 'nicole')
print(lista_de_chamada)
````

### Adicionando v√°rios elementos

`list.extend([elmt1, elemt2, elmt3, ...])`

````python
numeros_pares = [0,2,4,6,8]

numeros_impares = [1,3,5,7,9]

numeros_pares.extend(numeros_impares)

print(numeros_pares)
print(numeros_impares)
````

## ‚ûñ Removendo elementos

Tamb√©m √© poss√≠vel remover um elemento de uma lista, seja pela posi√ß√£o, seja pelo pr√≥prio elemento.

### Removendo um elemento por posi√ß√£o (`del`)

`del(lista[pos])`

> ‚ö†Ô∏è A fun√ß√£o `del()` n√£o √© uma fun√ß√£o embutida das listas e pode ser usada para remover qualquer elemento da mem√≥ria. **Use com cuidado!**

````python
convidados = ['Dennis', 'Malu', 'Aquiles']
print(convidados)
del(convidados[2])
print(convidados)
````

### Removendo o √∫ltimo elemento (ou algum elemento por posi√ß√£o)

`elem_removido = lista.pop()`

`elem_removido = lista.pop(pos)`

> üí° O m√©todo `pop()` sempre retorna o elemento removido

````python
convidados = ['Dennis', 'Malu', 'Aquiles']
print(convidados)
amigo_que_saiu_da_lista = convidados.pop(1)
print(amigo_que_saiu_da_lista)
print(convidados)
````

### Removendo elemento por busca

`lista.remove(elem)`

````python
lista_registros = [32163213,12321321,231231,51515,15151,3232]
print(lista_registros)
lista_registros.remove(51515)
print(lista_registros)
````

# üì∂ Ordena√ß√£o de listas

Muitas vezes, √© necess√°rio ter as lista ordenadas. Quando uma lista √© exta√≠da a partir da entrada de um usu√°rio ou alguma outra fonte, √© comum que ela venha fora de ordem. Para isso, podemos usar as seguintes fun√ß√µes:
- `sorted(lista)`
- `l.sort()`
- `l.reverse()`

> ‚ö†Ô∏è Para que essas fun√ß√µes funcionem adequadamente, √© necess√°rio que todos os elementos da lista sejam apenas **n√∫meros** ou **strings**, mas n√£o ambos ao mesmo tempo.

## Ordem alfab√©tica (visualiza√ß√£o)

```python
nova_lista = sorted(lista)
```

```python
nova_lista = sorted(lista, reverse=True)
```

> üí° A fun√ß√£o `sorted()` n√£o modifica a lista, mas retorna uma nova lista ordenada.


````python
lista_numeros = [50,6,2,9,54,61,99]
nova_lista = sorted(lista_numeros)
print(lista_numeros)
print(nova_lista)

nova_lista_decre = sorted(lista_numeros, reverse=True)
print(nova_lista_decre)
````

## Ordem alfab√©tica (definitiva)

```
lista.sort()
```

```python
lista.sort(reverse=True)
```

> üí° A fun√ß√£o `sort()` modifica, de forma definitiva, a posi√ß√£o dos elementos dentro da lista.

````python
convidados = ['zuleide', 'arthur','dennis']
print(convidados)
convidados.sort(reverse=True)
print(convidados)

````

## Invertendo a ordem de inser√ß√£o

```python
lista.reverse()
```

> üí° O `reverse()` n√£o usa nenhum algoritmo de ordena√ß√£o, apenas inverte a posi√ß√£o de todos os elementos.

````python
convidados = ['zuleide', 'arthur','dennis']
#convidados.reverse()
#print(convidados)

nova_lista = convidados[::-1]
print(convidados)
````

## C√≥pia de uma lista

Quando queremos modificar uma lista sem perder a ordena√ß√£o e os elementos originai, √© recomendado fazer uma c√≥pia da lista.

```python
l.copy()
l[:]
```

````python
lista_0_10 = list(range(0,10))

nova_lista_1 = lista_0_10

nova_lista_2 = lista_0_10[:]
nova_lista_3 = lista_0_10.copy()

lista_0_10.append(99)
nova_lista_1.append(0)
nova_lista_2.append(25)
nova_lista_3.append(50)

print(lista_0_10)

print("########")
print(nova_lista_1)
print(nova_lista_2)
print(nova_lista_3)

print("########")
print(lista_0_10)
print(nova_lista_1)
````

3. Considerando que as seguintes listas est√£o relacionadas pelo modelo e  consumo de combust√≠vel (*km/L*), exiba:

```python
modelos = ['Variant', 'Bras√≠lia', 'Chevette', 'Opala', 'Belina', 'Maverick']
consumos = [12.0, 9.3, 10.3, 7.3, 9.3, 9.1]
```

a) O modelo que consome mais combust√≠vel;<br/>
b) O modelo que consome menos combust√≠vel;<br/>

````python
modelos = ['Variant', 'Bras√≠lia', 'Chevette', 'Opala', 'Belina', 'Maverick']
consumos = [12.0, 9.3, 10.3, 7.3, 9.3, 9.1]

maior_consumo = max(consumos)
menor_consumo = min(consumos)
media = sum(consumos)/len(consumos)

print(f'Modelo que mais consome: {modelos[consumos.index(maior_consumo)]}')
print(f'Modelo que menos consome: {modelos[consumos.index(menor_consumo)]}')
print(f'M√©dia de consumo: {media}')
````
