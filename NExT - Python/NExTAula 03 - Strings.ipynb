# üî† `str`

Em Python, textos s√£o armazenados como uma sequ√™ncia de caracteres [Unicode](https://pt.wikipedia.org/wiki/Unicode) - uma extens√£o do [ASCII](https://www.ime.usp.br/~kellyrb/mac2166_2015/tabela_ascii.html). Essa segu√™ncia, conhecida como **String** √© representada como um objeto do tipo `str`.

Por ser uma sequ√™ncia, √© possivel acessar cada caracter atrav√©s de um √≠ndice. Al√©m disso, o tipo `str` fornece uma variedade de m√©todos de manipula√ß√£o desses caracteres.

√â poss√≠vel criar uma **string literal** (aquela que est√° escrita diretamente no c√≥digo) usando aspas **simples**, **duplas** ou _**"triplas"**_:

````python
# Aspas simples
txt1 = 'exemplo'

# Aspas duplas
txt2 = "outro exemplo"

# Aspas triplas*
txt3 = '''que permite que voc√™
fa√ßa quebras de linhas
durante o texto'''

print(txt1, txt2, txt3)
````

Outra forma de criar um string √© usando o m√©todo `str()`

````python
# convertendo int em str
num = str(10)
print(num)

# convertendo float em str
f = str(12.3)
print(f)

# convertendo bool em str
b = str(True)
print(b)
````

## üî£ Caracteres de Escape

S√£o chamados de **caracteres de escape** marcadores usados para acessar caracteres do c√≥digo Unicode. Para escreve-los, use `contra-barra + caractere`:

| sintaxe | descri√ß√£o |
|---------|-----------|
| `\n` | Quebra de linha |
| `\t` | Tabula√ß√£o |
| `\'` | Aspas simples |
| `\"` | Aspas duplas |
| `\\` | Contra-barra |

> Obs.: A interpreta√ß√£o dos caracteres de escape pode ser "desativada" usando Raw Strings, marcadas com um `r` antes das aspas:
```python
text = r'Exemplo de texto\nSem caracteres de escape'
```

````python
# caracteres de escape
texto = 'um \"exemplo\" \n\tde \'texto\' \\nova parte'
print(texto)

# r-strings
texto = r'um \"exemplo\" \n\tde \'texto\' \\nova parte'
print(texto)
````

Como as *strings* de Python s√£o cole√ß√µes de caracteres, quando passamos como par√¢metro um `str` para uma fun√ß√£o `len()`, ela retorna a quantade de caracteres do texto.

# üàÅ Indexa√ß√£o de Strings

<!--![Hello](https://developers.google.com/edu/python/images/hello.png)-->

|str|‚Üí|P|y|t|h|o|n|
|-|-|-|-|-|-|-|-|
|Index +|‚Üí|0|1|2|3|4|5|6|
|Index -|‚Üí|-6|-5|-4|-3|-2|-1|

- **Index**: Cada caractere dentro de uma string pode ser acessado pela sua posi√ß√£o (√≠ndice). Esse *index* pode ser positivo ou negativo.

```python
py = 'Python'
py[3] # h
py[-2] # o
```

- **Slicing**: Al√©m de um caractere espec√≠fico acessado pela posi√ß√£o, √© poss√≠vel acessar fazer o *fatiamento*. Fatiar permite que voc√™ obtenha uma **substring** da *string* original:

```python
txt[1:4] # caracteres da posi√ß√£o 1 (inclu√≠do) at√© 4 (exclu√≠do)
txt[2:5] # caracteres da posi√ß√£o 2 (inclu√≠do) at√© 5 (exclu√≠do)
txt[:5]  # caracteres da posi√ß√£o 0 (inclu√≠do) at√© 5 (exclu√≠do)
```

- **Slicing com passo**: Ainda existe uma outra nota√ß√£o que permite acessar *substrings* de uma posi√ß√£o at√© outra determinando o passo:

```python
txt[0:20:2] # caracteres da posi√ß√£o 0 (inclu√≠do) at√© 20 (exclu√≠do), a cada 2
```

> Obs.: *N√£o existe um tipo "caractere". Em Python um caracter √© uma string que possui tamanho 1.*

````python
txt = 'Ovos, bacon, salsicha e SPAM!'
print(txt)

# Posi√ß√£o por index
print(txt[0])
print(txt[1])
print(txt[-2])

# Slicing
print(txt[6:11])
print(txt[6:])
print(txt[:22])

# Slicing com passo
print(txt[6:11:3])

# Substring com passo negativo
print(txt[6:11:-1])
print(txt[10:5:-1])
````

````python
print(txt[::-1])
````

## *Ô∏è‚É£ Opera√ß√µes com Strings

>‚ö†Ô∏è Strings em Python s√£o imut√°veis, o que signifca que atribuir um valor a uma posi√ß√£o indexada ir√° resultar em um erro.

Os seguintes operadores podem ser usados com strings:

- `+` - Concatena√ß√£o
- `*` - Repeti√ß√£o

````python
# Concatena√ß√£o
nome = 'Rodolfo'
sobrenome = 'Oliveira'
nome_completo = nome + ' ' + sobrenome

print(nome_completo)

# Repeti√ß√£o
print('Nan\n' * 5, 'Batman')
print('-'*15)
print((sobrenome + '\n') * 10)
````

````python
nome[0] = 'S'
print(nome)
````

## `in`

Este operador permite avaliar se determinada substring existe dentro de uma string. A expre√ß√£o retorna `True` se existir.

```python
'nada' in 'S√≥ sei que nada sei'
```

> üí° *A combina√ß√£o com* `not in` *tamb√©m pode ser usada.*

## str.find()

Retorna a posi√ß√£o de uma determinada *substring* procurada.

Se o valor retornado for:
- `>= 0` : *Substring* encontrado numa posi√ß√£o espec√≠fica
- `< 0` : *Substring* n√£o encontrado

````python
email = 'erick.simoes@cesar@school'

print(email.find('@'))
print(email.find('simoes'))
print(email.find('z'))
````

````python
print(email.rfind('@'))
````

# `f-string`

> formatted string literals

As `f-strings` ou "*strings formatadas*" s√£o uma solu√ß√£o para quando precisamos exibir um texto que tem partes vari√°veis ou quando precisamos limitar a quantidade de caracteres de um n√∫mero.

```python
name = 'Erick'
print(f'Meu nome √© {name}')
```

Mais detalhes sobre f-Strings [aqui](https://realpython.com/python-f-strings/).

````python
pi = 3.14159

nome = 'Nicole'
altura = 1.59
golpe = 'mata-le√£o'

#saida = 'A monitora ' + nome + ' tem ' + str(altura) + ' de altura e sabe dar um perigoso ' + golpe

# adicionando vari√°veis
saida = f'A monitora {nome} tem {altura} de altura e sabe dar um perigoso {golpe}'
print(saida)

# alterando a quantidade de casas decimais de um n√∫mero:

print("O valor de pi √©: {}".format(pi))
print(f"O valor de pi √©: {pi:.2f}")
print(pi)

valor = 12
print(f"{valor:02}")

````

16. Leia uma cadeia de caracteres e converta todos os caracteres para mai√∫sculo **sem usar a fun√ß√£o `upper()`**. 

üí°Dica: subtraia 32 dos caracteres cujo codigo ASCII esta entre 65 e 90.

````python
t = input('Insira um texto: ')

for l in t:
  if 97 <= ord(l) <= 122:
    print(chr(ord(l) - 32), end='')
  else:
    print(l, end='')
````

````python
# v√°lido:   1 10 -1 -10 10.2 1.1 -1.0
# inv√°lido: 1- 1-0 .1 1.
# If composto
num = '-.12'

if num.isdigit():
  print('Valor inteiro positivo')
elif num.find('-') == 0 and num[1:].isdigit():
  print('Valor inteiro negativo')
elif num.count('.') == 1 and num.count('-') == 0 and 1 <= num.find('.') < len(num) -1:
  print('Valor real positivo')
elif num.count('.') == 1 and num.count('-') == 1 and num.find('-') == 0 and 1 < num.find('.') < len(num) -1:
  print('Valor real negativo')
else:
  print('-- Valor inv√°lido --')
````
