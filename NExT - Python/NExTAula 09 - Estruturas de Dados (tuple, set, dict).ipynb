# Estruturas de Dados

Sempre que precisamos armazenar v√°rios dados em uma √∫nica vari√°vel se faz necess√°rio usar algum tipo de dado **Container**. Em Python, os principais s√£o:
- `list`;
- `tuple`;
- `set`;
- `dict`;

Cada containers, ou tipos de cole√ß√£o, tem uma caracter√≠stica espec√≠fica e um uso diferente. Nesta aula vamos conhecer um pouco mais sobre eles.

## ü™® `tuple`

Ao usar listas, h√° bastante liberdade para alterar seus itens. Em determinadas situ√ß√µes, essas listas n√£o deveriam ser alteradas, tais como:
- Dias da semana;
- Coordenadas de um pr√©dio;
- Letras de um alfabeto.

> ### Caracter√≠sticas:
- Imut√°veis;
- Ordenados/index√°vel;
- Permite elementos duplicados.


```python
t = ('domingo', 'segunda', 'ter√ßa', 'quarta', 'quinta', 'sexta', 's√°bado')
```

> üí° Obs.: *O acesso aos elementos de uma tupla √© feito da mesma forma que com listas*

````python
# Cria√ß√£o de uma Tupla
dias_uteis = ('segunda', 'ter√ßa', 'quarta', 'quinta', 'sexta')
naipes = 'espadas', 'paus', 'copas', 'ouro'

print(dias_uteis)
print(type(naipes))

# Acesso aos elementos de uma Tupla
print(dias_uteis[:2])

# len()
print(len(dias_uteis))

# tupla.index(item)
print(dias_uteis.index('sexta'))
print('domingo' in dias_uteis)

# tupla.count(item)
print(naipes.count('ouro'))

# Imutabilidade
dias_uteis[0] = 'domingo'
````

````python
entrada = tuple(input('Insira os nomes das pessoas: ').split())

print(type(entrada))
print(entrada)

#tupla_entrada = tuple(entrada)

#print(type(tupla_entrada))
#print(tupla_entrada)
````

## üíé `set`

Tanto em **listas**, quanto em **tuplas**, √© poss√≠vel ter elementos repetidos:

```python
t = (0, 1, 1, 2, 3, 5)
l = [0, 1, 1, 2, 3, 5]
```

Quanto houver uma situa√ß√£o onde voc√™ n√£o deva armazenar itens repetidos numa lista, √© poss√≠vel usar um `set`:

```python
s1 = set()
s.add(1)
s.add(2)
s.add(1)

s2 = {'bacon', 'banana', 'spam', 'ovos', 'spam', 'salsicha', 'spam'}
```

> ### Caracter√≠sticas:
- Imut√°veis (mas podem ser adicionados e removidos);
- N√£o Ordenados/ n√£o index√°vel;
- N√£o Permite elementos duplicados.

````python
# criar um set
monitores = set()
ingredientes = {'bacon', 'ovos', 'spam', 'spam', 'spam'}

print(type(monitores))
print(monitores)

#print(ingredientes[0]) # erro!

# adicionar elementos
monitores.add('Lev')
monitores.add('Jo√£o')
monitores.add('Patrick')
monitores.add('Lev')
monitores.add('Jo√£o')
monitores.add('Patrick')
monitores.add('Lev')
monitores.add('Jo√£o')
monitores.add('Patrick')

print(monitores)

nums = [1, 2, 3, 1, 4, 5, 5, 5, 5, 7, 8, 9, 1]
print(len(set(nums)))
nums_n_repetidos = set(nums)

print(nums_n_repetidos)

# remover elementos
print(ingredientes)
ingredientes.remove('spam')
print(ingredientes)
````

````python
ingredientes = {'bacon', 'ovos', 'spam', 'spam', 'spam'}

for item in ingredientes:
  print(item)
````

````python
n = set([1, 1.0, 'omena', False, 0])

print(n)
````

## üîë Dicion√°rios

Para criar estruturas de dados que s√£o mapeados por valores (e n√£o pelo indice, como nas listas) √© poss√≠vel usar um dicion√°rio.

> **Chave** x **Valor**

Obs.: *O mapeamento √© feito pela chave (e n√£o pela ordem)*

Exemplos de uso pr√°tico:
- Dicion√°rios;
- CEP;

```python
dicionario = {'chave': 'valor1', 'chave2': 'valor2'}
```

> ### Caracter√≠sticas:
- Mut√°veis;
- Ordenados/index√°vel;
- Permite valores duplicados (mas n√£o a chave).

````python
# Criando um dicion√°rio
cep = {'5102154': ('Avenida cais da rua', 'Recife Antigo', 'PE'),
       '5355647': 'Rua da saudade que n√£o vai',
       '1125478': 'Rua da avendia da estrada'
       }
tradutor = {'dog':'c√£o', 'lion':'le√£o', 'bird':'p√°ssaro'}
agenda = {'m√£e':(123456, 789456, 88774455), 'pai':(753159,)}

# Acessando elementos de um dicion√°rio
print(tradutor['dog'])
print(tradutor['bird'])

print(len(agenda['m√£e']))
print(agenda['m√£e'][-1])

# Adi√ß√£o de itens por atribui√ß√£o
pokemon = dict()

pokemon['Charmander'] = 'fogo'
pokemon['Psyduck'] = 'psiquico', 'agua'

print(pokemon)

pokemon['Psyduck'] = 'agua'

#nome = input('Insira o nome: ')
#tipo = input('Qual o tipo? ')

#pokemon[nome] = tipo

print(pokemon)

print(len(agenda['m√£e']))
print(len(agenda['pai']))

# Adi√ß√£o de itens por update
pokemon.update({'Pikachu':'El√©trico'})
print(pokemon)

# Removendo itens
#pokemon.popitem()
print(pokemon)

#pokemon.pop('Psyduck')
#del pokemon['Psyduck']
print(pokemon)

# esvaziando um dicion√°rio
#pokemon.clear()
print(pokemon)
````

````python
pokemons = dict()

for _ in range(5):
  key = input('Insira o nome do Pokemon: ')
  value = input('Qual o tipo desse Pokemon? ')

  pokemons[key] = value

print(pokemons)
````

### üóùÔ∏è Alguns m√©todos √∫teis de dicion√°rios

- `dict.keys()` - Apresenta todas as chaves do dicion√°rio
- `dict.values()` - Apresenta todos os valores do dicion√°rio
- `dict.items()` - `dict_keys()` + `dict_values()`

````python
# keys()
print(pokemon.keys())
#print(list(pokemon.keys())[:2])

# values()
print(pokemon.values())

# items()
print(pokemon.items())

# for
for elem in pokemon:
  print(elem)

print('*'*20)

for elem in pokemon.values():
  print(elem)

print('*'*20)

for elem in pokemon.items():
  print(elem)
  print(f'A chave √© {elem[0]} e o valor √© {elem[1]}')

print('*'*20)

for nome, tipo in pokemon.items():
  print(f'O pokemon {nome} √© do tipo {tipo}')
````

2. Crie um programa com duas fun√ß√µes. A primeira deve pedir um CEP e o endere√ßo do usu√°rio e armazena-lo em um dicion√°rio. A segunda deve pesquisar um endere√ßo, buscando pelo CEP informado.

üèÜ Desafio: Armazene o dicion√°rio em um arquivo.

````python
ceps = {}

def inclui_endereco():
  cep = input('Insira um CEP: ')
  endereco = input(f'Insira um endere√ßo do CEP {cep}: ')

  ceps.update({cep:endereco})


def busca_endereco():
  cep = input('Insira um CEP que voc√™ deseja buscar: ')

  print(ceps.get(cep, 'CEP n√£o encontrado'))

while True:
  entrada = input('O que voc√™ deseja fazer:\n\t1 - Inserir um CEP\n\t2 - Buscar um CEP\n\t3 - Sair\n')

  if entrada == '1':
    inclui_endereco()
  elif entrada == '2':
    busca_endereco()
  elif entrada == '3':
    print('Volte sempre!')
    break
  else:
    print('Comando inv√°lido, tente novamente...')

  print(ceps)
````
